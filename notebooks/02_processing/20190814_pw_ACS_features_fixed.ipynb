## Feature Engineering

```python
df.sort_values(['district_id', 'year'], inplace=True)
```

```python
df['M75_above'] = df['Male 75 to 79 years'] + \
    df['Male 80 to 84 years'] + df['Male 85 years and over']
df['F75_above'] = df['Female 75 to 79 years'] + \
    df['Female 80 to 84 years'] + df['Female 85 years and over']

df['M60_74'] = df['Male 70 to 74 years'] + df['Male 67 to 69 years'] + \
    df['Male 65 and 66 years'] + \
    df['Male 60 and 61 years'] + df['Male 62 to 64 years']
df['F60_74'] = df['Female 70 to 74 years'] + df['Female 67 to 69 years'] + \
    df['Female 65 and 66 years'] + \
    df['Female 60 and 61 years'] + df['Female 62 to 64 years']

df['M45_59'] = df['Male 45 to 49 years'] + \
    df['Male 50 to 54 years'] + df['Male 55 to 59 years']
df['F45_59'] = df['Female 45 to 49 years'] + \
    df['Female 50 to 54 years'] + df['Female 55 to 59 years']

df['M30_44'] = df['Male 30 to 34 years'] + \
    df['Male 35 to 39 years'] + df['Male 40 to 44 years']
df['F30_44'] = df['Female 30 to 34 years'] + \
    df['Female 35 to 39 years'] + df['Female 40 to 44 years']

df['M18_29'] = df['Male 18 and 19 years'] + df['Male 20 years'] + df['Male 21 years'] + df['Male 22 to 24 years'] + \
    df['Male 25 to 29 years']
df['F18_29'] = df['Female 18 and 19 years'] + df['Female 20 years'] + df['Female 21 years'] + df['Female 22 to 24 years'] + \
    df['Female 25 to 29 years']

df['M18_below'] = df['Total Male'] - df['M18_29'] - \
    df['M30_44'] - df['M45_59'] - df['M60_74'] - df['M75_above']
df['F18_below'] = df['Total Female'] - df['F18_29'] - \
    df['F30_44'] - df['F45_59'] - df['F60_74'] - df['F75_above']

df['M18_above'] = df['Total Male'] - df['M18_below']
df['F18_above'] = df['Total Female'] - df['F18_below']

df['M25_above'] = df['Total Male'] - df['M18_below'] - \
    (df['Male 18 and 19 years'] + df['Male 20 years'] +
     df['Male 21 years'] + df['Male 22 to 24 years'])

df['F25_above'] = df['Total Female'] - df['F18_below'] - \
    (df['Female 18 and 19 years'] + df['Female 20 years'] +
     df['Female 21 years'] + df['Female 22 to 24 years'])
```

```python
df['Mpop_share'] = df['Total Male']/df['TOTAL POPULATION']
df['Fpop_share'] = df['Total Female']/df['TOTAL POPULATION']
```

```python
df['M18_below_share'] = df['M18_below'] / df['Total Male']
df['F18_below_share'] = df['F18_below'] / df['Total Female']
df['18_below_share'] = (df['M18_below'] + df['F18_below'])/df['TOTAL POPULATION']

df['M18_above_share'] = df['M18_above'] / df['Total Male']
df['F18_above_share'] = df['F18_above'] / df['Total Female']
df['18_above_share'] = (df['M18_above'] + df['F18_above'])/df['TOTAL POPULATION']

df['M18_29_share'] = df['M18_29'] / df['Total Male']
df['F18_29_share'] = df['F18_29'] / df['Total Female']
df['18_29_share'] = (df['M18_29'] + df['F18_29'])/df['TOTAL POPULATION']

df['M30_44_share'] = df['M30_44'] / df['Total Male']
df['F30_44_share'] = df['F30_44'] / df['Total Female']
df['30_44_share'] = (df['M30_44'] + df['F30_44'])/df['TOTAL POPULATION']

df['M45_59_share'] = df['M45_59'] / df['Total Male']
df['F45_59_share'] = df['F45_59'] / df['Total Female']
df['45_59_share'] = (df['M45_59'] + df['F45_59'])/df['TOTAL POPULATION']

df['M60_74_share'] = df['M60_74'] / df['Total Male']
df['F60_74_share'] = df['F60_74'] / df['Total Female']
df['60_74_share'] = (df['M60_74'] + df['F60_74'])/df['TOTAL POPULATION']

df['M75_above_share'] = df['M75_above'] / df['Total Male']
df['F75_above_share'] = df['F75_above'] / df['Total Female']
df['75_above_share'] = (df['M75_above'] + df['F75_above_share'])/df['TOTAL POPULATION']
```

```python
df['white_share'] = df['White alone'] / df['TOTAL POPULATION']
df['black_share'] = df['Black or African American alone'] / df['TOTAL POPULATION']
df['asian_share'] = df['Asian alone'] / df['TOTAL POPULATION']
df['hispanic_share'] = df['Hispanic or Latino'] / df['TOTAL POPULATION']
df['otherrace_share'] = (df['TOTAL POPULATION']-df['White alone']-df['Black or African American alone']-df['Asian alone']) / df['TOTAL POPULATION']
df['native_share'] = df['Native'] / df['TOTAL POPULATION']
df['nativeinstate_share'] = df['Native Born in state of residence'] / df['TOTAL POPULATION']
df['nativeoutofstate_share'] = df['Native Born in other state'] / df['TOTAL POPULATION']
df['foreignborn_share'] = df['Foreign born'] / df['TOTAL POPULATION']
```

```python
df['HS_below_share'] = df['Less than high school graduate'] / df['population 25 and over']
df['HS_share'] = df['High school graduate'] / df['population 25 and over']
df['somecollege_share'] = df['Some college or associate degree'] / df['population 25 and over']
df['college_share'] = df['Bachelors degree'] / df['population 25 and over']
df['graddeg_share'] = df['Graduate or professional degree'] / df['population 25 and over']

df['samehouse_share'] = df['Same house 1 year ago'] / df['TOTAL POPULATION']
df['samecounty_share'] = df['Moved within same county'] / df['TOTAL POPULATION']
df['samestate_share'] = df['Moved from different county'] / df['TOTAL POPULATION']
df['diffstate_share'] = df['Moved from different state'] / df['TOTAL POPULATION']
df['liveabroad_share'] = df['Moved from abroad'] / df['TOTAL POPULATION']
```

```python
df['ptransport_share'] = df['Public transportation']/df['Workers 16 and over']
df['walktowork_share'] = df['Walked']/df['Workers 16 and over']
df['workathome_share'] = df['Worked at home']/df['Workers 16 and over']

df['inschool_share'] = df['Enrolled in school']/df['TOTAL POPULATION']
df['incollege_share'] = df['Enrolled in college']/(df['M18_above'] + df['F18_above'])
df['ingradschool_share'] = df['Enrolled in school Graduate professional']/(df['M18_above'] + df['F18_above'])
```

```python
df['M_college_share'] = df['Male Bachelors']/df['M25_above']
df['M_graddeg_share'] = (df['Male Master'] + df['Male Professional school'])/df['M25_above']
df['M_phd_share'] = df['Male Doctorate']/df['M25_above']

df['F_college_share'] = df['Female Bachelor']/df['F25_above']
df['F_graddeg_share'] = (df['Female Master'] + df['Female Professional school'])/df['F25_above']
df['F_phd_share'] = df['Female Doctorate']/df['F25_above']
```

```python
df['poverty_share'] = df['Income below poverty'] / df['TOTAL POPULATION']
df['hhinc_10k_less_share'] = df['Income Less than 10000'] / df['Total Household Income']
```

```python
df['hhinc_30k_less_share'] = (df['Income Less than 10000'] + df['Income 10000 to 14999']\
                              + df['Income 15000 to 19999'] +
                              df['Income 20000 to 24999'] + df['Income 25000 to 29999'])/\
df['Total Household Income']
```

```python
df['hhinc_50k_less_share'] = (df['Income Less than 10000'] + df['Income 10000 to 14999']\
                              + df['Income 15000 to 19999'] +
                              df['Income 20000 to 24999'] + df['Income 25000 to 29999']\
                              + df['Income 30000 to 34999'] + df['Income 35000 to 39999']\
                              + df['Income 40000 to 44999'] +\
                              df['Income 45000 to 49999'])/ df['Total Household Income']
```

```python
df['hhinc_75k_more_share'] = (df['Income 75000 to 99999'] + df['Income 100000 to 124999']\
                              + df['Income 125000 to 149999'] + df['Income 150000 to 199999']\
                              + df['Income 200000 or more'])/df['Total Household Income']
```

```python
df['hhinc_100k_more_share'] = (df['Income 100000 to 124999']\
                              + df['Income 125000 to 149999'] + df['Income 150000 to 199999']\
                              + df['Income 200000 or more'])/df['Total Household Income']
```

```python
df['hhinc_125k_more_share'] = (df['Income 125000 to 149999'] + df['Income 150000 to 199999']\
                              + df['Income 200000 or more'])/df['Total Household Income']
```

```python
df['hhinc_150k_more_share'] = (df['Income 150000 to 199999']\
                              + df['Income 200000 or more'])/df['Total Household Income']
```

```python
df['hhinc_200k_more_share'] = df['Income 200000 or more']/df['Total Household Income']
```

```python
df['veteran_share'] = df['Veteran']/(df['M18_above'] + df['F18_above'])
df['LFP_share'] = df['In labor force']/df['Population 16 and over']
df['Unemp_rate'] = df['Civilian Unemployed']/df['Civilian in labor force']
df['armedforce_share'] = df['In Armed Forces']/df['In labor force']
df['naturalresourcesemp_share'] = (df['Male natural resources']+df['Female natural resources'])\
    / df['Fulltime civilian employed 16 and over']
df['bluecollaremp_share'] = (df['Male production transportation and material moving']
                             + df['Female production transportation and material moving'])\
    / df['Fulltime civilian employed 16 and over']
```

```python
df['vacanthousing_share'] = df['Vacant']/ df['TOTAL HOUSING UNITS']
df['renter_share'] = df['Renter occupied']/df['Occupied Housing units']
df['mortgage_share'] = df['Housing units with a mortgage']/df['Owner occupied housing units']

df['housevalue_10k_less_share'] = df['Housing Value Below 10000']/df['Total owner occupied housing units']

df['housevalue_50k_less_share'] = (df['Housing Value Below 10000'] + df['Housing Value 10000 to 14999'] +\
df['Housing Value 15000 to 19999']+df['Housing Value 20000 to 24999']+df['Housing Value 25000 to 29999']\
+df['Housing Value 30000 to 34999']+df['Housing Value 35000 to 39999']+df['Housing Value 40000 to 49999'])\
/df['Total owner occupied housing units']

df['housevalue_100k_less_share'] = (df['Housing Value Below 10000'] + df['Housing Value 10000 to 14999'] +\
df['Housing Value 15000 to 19999']+df['Housing Value 20000 to 24999']+df['Housing Value 25000 to 29999']\
+df['Housing Value 30000 to 34999']+df['Housing Value 35000 to 39999']+df['Housing Value 40000 to 49999']\
+df['Housing Value 50000 to 59999'] + df['Housing Value 60000 to 69999'] +\
df['Housing Value 70000 to 79999']+df['Housing Value 80000 to 89999']+df['Housing Value 90000 to 99999'])\
/df['Total owner occupied housing units']

df['housevalue_300k_more_share'] = (df['Housing Value 300000 to 399999'] + df['Housing Value 400000 to 499999'] +\
df['Housing Value 500000 to 749999']+df['Housing Value 750000 to 999999']+df['Housing Value 1000000 to 1499999'])\
/df['Total owner occupied housing units']

df['housevalue_500k_more_share'] = (df['Housing Value 500000 to 749999']+df['Housing Value 750000 to 999999']\
                                    +df['Housing Value 1000000 to 1499999'])/df['Total owner occupied housing units']

df['housevalue_750k_more_share'] = (df['Housing Value 750000 to 999999']\
                                    +df['Housing Value 1000000 to 1499999'])/df['Total owner occupied housing units']
df['housevalue_1m_more_share'] = df['Housing Value 1000000 to 1499999']/df['Total owner occupied housing units']
```
